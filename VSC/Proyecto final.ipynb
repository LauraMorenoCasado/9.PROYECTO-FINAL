```python
import pandas as pd
import numpy as np
```

```python
#Ahora vamos a proceder a importar los ficheros en formato excel procedentes de la empresa, sólo las hojas excel que nos interesan analizar.
FT_1 = pd.read_excel("C:./Fábrica de Toner.xlsx", sheet_name=0, index_col=0)
FT_2 = pd.read_excel("C:./Fábrica de Toner.xlsx", sheet_name=2, index_col=0)
GTH_1 = pd.read_excel("C:./Gestión de Talento Humano.xlsx", sheet_name=0, index_col=0)
GTH_2 = pd.read_excel("C:./Gestión de Talento Humano.xlsx", sheet_name=1, index_col=0)
Licitaciones = pd.read_excel("C:./Licitaciones.xlsx", sheet_name=0, index_col=0)
SAT_1 = pd.read_excel("C:./SAT.xlsx", sheet_name=2, index_col=0)
SAT_2 = pd.read_excel("C:./SAT.xlsx", sheet_name=3, index_col=0)
TRÁFICO_1 = pd.read_excel("C:./TRÁFICO.xlsx", sheet_name=0, index_col=0)
TRÁFICO_2 = pd.read_excel("C:./TRÁFICO.xlsx", sheet_name=2, index_col=0)
Ventas = pd.read_excel("C:./Ventas.xlsx", sheet_name=0, index_col=0)
```

```python
#Ahora procedemos a printear los dataframes para ver que se han importado correctamente.
print(FT_1.head())
print(FT_2.head())
print(GTH_1.head())
print(GTH_2.head())
print(Licitaciones.head())
print(SAT_1.head())
print(SAT_2.head())
print(TRÁFICO_1.head())
print(TRÁFICO_2.head())
print(Ventas.head())
```

# Patrones, identificación, anomalías, verificación y suposiciones de datos

```python
#Comprobamos los tipos de datos de cada dataframe.
print(FT_1.dtypes)
print(FT_2.dtypes)
print(GTH_1.dtypes)
print(GTH_2.dtypes)
print(Licitaciones.dtypes)
print(SAT_1.dtypes)
print(SAT_2.dtypes)
print(TRÁFICO_1.dtypes)
print(TRÁFICO_2.dtypes)
print(Ventas.dtypes)
#La gran mayoría de los datos están en el formato correcto, pero hay algunos que no.
#Procedemos a corregir los tipos de datos incorrectos.
Ventas['Fecha'] = pd.to_datetime(Ventas['Fecha'], errors='coerce')
#Volvemos a comprobar los tipos de datos para asegurarnos de que todo está correcto.
print(Ventas.dtypes)
```

```python
#Comprobamos si hay columnas completamente vacías en los dataframes.
print(FT_1.columns[FT_1.isnull().all()])
print(FT_2.columns[FT_2.isnull().all()])
print(GTH_1.columns[GTH_1.isnull().all()])
print(GTH_2.columns[GTH_2.isnull().all()])
print(Licitaciones.columns[Licitaciones.isnull().all()])
print(SAT_1.columns[SAT_1.isnull().all()])
print(SAT_2.columns[SAT_2.isnull().all()])
print(TRÁFICO_1.columns[TRÁFICO_1.isnull().all()])
print(TRÁFICO_2.columns[TRÁFICO_2.isnull().all()])
print(Ventas.columns[Ventas.isnull().all()])
#No hay columnas completamente vacías en ninguno de los dataframes.
```

```python
#Comprobamos si hay valores duplicados en los dataframes.
print(FT_1.duplicated().sum())
print(FT_2.duplicated().sum())
print(GTH_1.duplicated().sum())
print(GTH_2.duplicated().sum())
print(Licitaciones.duplicated().sum())
print(SAT_1.duplicated().sum())
print(SAT_2.duplicated().sum())
print(TRÁFICO_1.duplicated().sum())
print(TRÁFICO_2.duplicated().sum())
print(Ventas.duplicated().sum())
#No hay valores duplicados en ninguno de los dataframes.
```

```python
#Comprobamos si hay valores nulos en los dataframes.
print(FT_1.isnull().sum())
print(FT_2.isnull().sum())
print(GTH_1.isnull().sum())
print(GTH_2.isnull().sum())
print(Licitaciones.isnull().sum())
print(SAT_1.isnull().sum())
print(SAT_2.isnull().sum())
print(TRÁFICO_1.isnull().sum())
print(TRÁFICO_2.isnull().sum())
print(Ventas.isnull().sum())
#No hay valores nulos en ninguno de los dataframes.
```

```python
#Comprobamos si hay valores nan en los dataframes.
print(FT_1.isna().sum())
print(FT_2.isna().sum())
print(GTH_1.isna().sum())
print(GTH_2.isna().sum())
print(Licitaciones.isna().sum())
print(SAT_1.isna().sum())
print(SAT_2.isna().sum())
print(TRÁFICO_1.isna().sum())
print(TRÁFICO_2.isna().sum())
print(Ventas.isna().sum())
#No hay valores nan en ninguno de los dataframes.
```

```python
#Como hemos observado que hay valores nan, vamos a proceder a cambiarlos en las columnas categóricas por "Desconocido".
categorical_columns = Licitaciones.select_dtypes(include=['object']).columns
for column in categorical_columns:
    Licitaciones[column] = Licitaciones[column].fillna('Desconocido')
```

```python
# 6. Análisis estadísticos de los principales atributos numéricos
print("Medias de atributos numéricos:")
print(FT_1.describe().loc['mean'])
print(FT_2.describe().loc['mean'])
print(GTH_1.describe().loc['mean'])
print(GTH_2.describe().loc['mean'])
print(Licitaciones.describe().loc['mean'])
print(SAT_1.describe().loc['mean'])
print(SAT_2.describe().loc['mean'])
print(TRÁFICO_1.describe().loc['mean'])
print(TRÁFICO_2.describe().loc['mean'])
print(Ventas.describe().loc['mean'])

print("Medianas de atributos numéricos:")
print(FT_1.median(numeric_only=True))
print(FT_2.median(numeric_only=True))
print(GTH_1.median(numeric_only=True))
print(GTH_2.median(numeric_only=True))
print(Licitaciones.median(numeric_only=True))
print(SAT_1.median(numeric_only=True))
print(SAT_2.median(numeric_only=True))
print(TRÁFICO_1.median(numeric_only=True))
print(TRÁFICO_2.median(numeric_only=True))
print(Ventas.median(numeric_only=True))

print("Desviaciones estándar de atributos numéricos:")
print(FT_1.std(numeric_only=True))
print(FT_2.std(numeric_only=True))
print(GTH_1.std(numeric_only=True))
print(GTH_2.std(numeric_only=True))
print(Licitaciones.std(numeric_only=True))
print(SAT_1.std(numeric_only=True))
print(SAT_2.std(numeric_only=True))
print(TRÁFICO_1.std(numeric_only=True))
print(TRÁFICO_2.std(numeric_only=True))
print(Ventas.std(numeric_only=True))

#Las medias, medianas y desviaciones estándar de los atributos numéricos parecen estar dentro de rangos razonables.
#No hay valores atípicos evidentes en los atributos numéricos.
```
